```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import pandas as pd
```

```python
X = np.array([130,250,190,300,210,220,170])
Y = np.array([16.3,10.2,11.1,7.1,12.1,13.2,14.2])
X = X[:, np.newaxis]

lin_model = linear_model.LinearRegression()
lin_model.fit(X,Y)
```

```python
print(lin_model.coef_) # 기울기
print(lin_model.intercept_) #y 절편
print(lin_model.score(X,Y)) #예측점수
```

```python
lin_model.predict([[270]])[0]  ##반드시 2차원으로 예측해야함. 그리고 [0] 해줘야 float로 반환됨
```

```python
A = np.array([130,250,190,300,210,220,170])
B = np.array([1900,2600,2200,2900,2400,2300,2100])
Y = np.array([16.3,10.2,11.1,7.1,12.1,13.2,14.2])

df = pd.DataFrame([A,B,Y],index=['horse','weight','eff'])
df = df.T
```

```python
model2 = linear_model.LinearRegression()
model2.fit(df[['horse','weight']],df[['eff']])
```

```python
print(model2.coef_[0]) # 기울기
print(model2.intercept_[0]) #y 절편
print(model2.score(df[['horse','weight']],df[['eff']])) #예측점수
```
